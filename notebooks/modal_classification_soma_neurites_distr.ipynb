```r
df<- read.csv("soma_neurites_distr.csv")
df<-droplevels(df)
df
```

```r

type_list <- c()
skew_list <- c()
for (i in 1:3163 )
{
a1 <- toString(df$freq_vector[i])
a1 <- as.character(a1) ;
#length(a1)
a2 <- substring(a1,2,nchar(a1)-1)

a3<- as.numeric(unlist(strsplit(a2,",")))


type_list <- c(type_list,ajus(a3)$type)
skew_list <- c(skew_list,ajus(a3)$skew)
    
}
```

```r
df2<-df
df2$modal_classification <- type_list
df2$skew_value <- skew_list
```
